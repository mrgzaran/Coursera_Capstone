#### Get data from url and transform into a pandas dataframe

```python
import numpy as np # library to handle data in a vectorized manner

import pandas as pd # library for data analsysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import json # library to handle JSON files

!conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans

!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library

print('Libraries imported.')
```

```python
url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
print('url done')
```

```python
soup = BeautifulSoup(url,'lxml')
print(soup.prettify())
```

```python
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

```python
df_toronto1=pd.read_html(tab)
df_toronto2=df_toronto1[0]
df_toronto2.head()
```

#### Clean and process data

```python
# Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.

df_toronto3 = df_toronto2[df_toronto2.Borough != 'Not assigned']

# More than one neighborhood can exist in one postal code area. Combine them into one row with the neighborhoods separated with a comma.

df_toronto3=df_toronto3.groupby(['Postal Code', 'Borough'], sort=False).agg( ', '.join)
df_toronto3.reset_index(inplace=True)

# If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.

df_toronto3['Neighbourhood'] = np.where(df_toronto3['Neighbourhood'] == 'Not assigned',df_toronto3['Borough'], df_toronto3['Neighbourhood'])

df_toronto3.head(20)
```

```python
# Shape
df_toronto3.shape
```

## Assignment 2
Now that you have built a dataframe of the postal code of each neighborhood along with the borough name and neighborhood name, in order to utilize the Foursquare location data, we need to get the latitude and the longitude coordinates of each neighborhood.

```python
# get latitude and longitude from the cvs file 
latlon = pd.read_csv('https://cocl.us/Geospatial_data')
latlon.head()
```

```python
# Merge df_toronto3 and latlon dataframes+
df_toronto4 = pd.merge(df_toronto3,latlon,on='Postal Code')
df_toronto4.head(10)
```

## Assignment 3
Explore and cluster the neighborhoods in Toronto. Make sure to:

```python
# set number of clusters
k = 5

toronto_grouped_clustering = df_toronto4.drop(['Postal Code','Borough','Neighbourhood'],1)

# run k-means clustering
kmeans = KMeans(n_clusters=k, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

#### Create a new dataframe that includes the clusters

```python
# add clustering labels
df_toronto4.insert(0, 'Cluster Labels', kmeans.labels_)
df_toronto4.head()
```

#### Create map

```python
# get longitude and latitude of Toronto
Latitude = 43.6529
Longitude = -79.3849
```

```python
# create map
map_clusters = folium.Map(location=[Latitude, Longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(df_toronto4['Latitude'], df_toronto4['Longitude'], df_toronto4['Neighbourhood'], df_toronto4['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=7,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
